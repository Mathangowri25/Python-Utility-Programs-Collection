## List Operation in lambda Function

```python
fun1 = lambda a:sum(a)
fun2 = lambda a:max(a)
fun3 = lambda a:min(a)
a = [10,20,30,40,50]
print(fun1(a))
print(fun2(a))
print(fun3(a))
```

```python
fun2 = lambda b:sorted(b)
b = [40,10,20,30]
fun2(b)
```

```python
fun3 = lambda c : sum([i for i in range(1,c+1)])
c = 10
fun3(c)
```

```python
fun4 = lambda c : "even" if c%2==0 else "odd"
c = 10
fun4(c)
```

```python
fun5 = lambda c:c[::-1]
c = "Tamilvendhen"
fun5(c)
```

```python
fun6 = lambda d:list(map(int,d**2))
d = np.array([10,20,30,40,50])
fun6(d)
```

```python
fun6 = lambda d:list(map(int,d**2))
d = np.array([10,20,30,40,50])
fun6(d)
```

## String method of lambda function

```python
fun3 = lambda c:b.split("-")[-1]
b = "pinto-cse-10,20,30"
fun3(b)
```

```python
fun3 = lambda c:b.split("-")[-1].split(",")
b = "pinto-cse-10,20,30"
fun3(b)
```

```python
# fun3 = lambda c:[int(d) for d in b.split("-")[-1].split(",")]
# b = "pinto-cse-10,20,30"
# fun3(b)

# or

fun3 = lambda c:list(map(int, b.split("-")[-1].split(",")))
b = "pinto-cse-10,20,30"
fun3(b)
```

```python
fun3 = lambda c:sum([int(d) for d in b.split("-")[-1].split(",")])
b = "pinto-cse-10,20,30"
fun3(b)
```

```python
fun3 = lambda c:",".join(map(str,b.split(",")[-4].split("-")))
fun4 = lambda c:b.split(",")[-4].split("-")
b = "pinto-cse,10,20,30"
print(fun3(b))
print(fun4(b))
```
